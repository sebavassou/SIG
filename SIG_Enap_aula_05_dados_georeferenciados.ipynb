### Transformações de Sistemas de Referências

Observações:
1.   O MAPA do **Leaflet** trabalha com SRC projetado (EPSG 3395 - WGS 84 / World Mercator)
2.   As coordenadas do **Leaflet** devem ser passada com o SRC Geográfico (**EPSG 4326** - WGS84)  
    2.1    O centro do mapa: Map( center=(Lat, Long))  
    2.2    As coordenadas das camadas DEVEM ser Geográficas (converter o SRC)

```python
import json
from ipyleaflet import Map, GeoJSON, Polygon as Polygon_ipyleaflet
from shapely.geometry import Polygon as Polygon_shapely

from osgeo import ogr, osr
ogr.UseExceptions()

def createLayerForLeaflet(ogr_geometry):
    geom = ogr_geometry.Clone()
    id_sr_source = geom.GetSpatialReference().GetAuthorityCode(None)
    id_sr_4326 = '4326'
    if not id_sr_source == id_sr_4326:
        # Transform
        sr_epsg_4326 = osr.SpatialReference()
        sr_epsg_4326.ImportFromEPSG( int( id_sr_4326 ) )
        result = geom.TransformTo( sr_epsg_4326 )

        if not result == ogr.OGRERR_NONE:
            msg = f"Error Geometry transform: '{id_sr_source}' to '{id_sr_4326}'"
            raise ValueError('Error transform ')

    geojson_data = {
        'type': 'Feature',
        'geometry': json.loads( geom.ExportToJson() ),
        'properties': {}
    }
    geom = None
    return GeoJSON(
        data=geojson_data,
        style={
            'color': 'purple',
            'opacity': 1,
            'weight': 1.9,
            'dashArray': '9',
            'fillOpacity': 0.4
        }
    )

def createBrasiliaSquareForLeaflet():
    polygon = Polygon_shapely([
        (-47.95, -15.75),
        (-47.95, -15.85),
        (-47.80, -15.85),
        (-47.80, -15.75),
        (-47.95, -15.75)
    ])
    center = ( polygon.centroid.y, polygon.centroid.x )
    coords = [[y, x] for x, y in polygon.exterior.coords]
    #
    layer = Polygon_ipyleaflet(
        locations=coords,
        color="blue",
        fill_color="blue",
        fill_opacity=0.4
    )
    return ( layer, center )

# Imovel em SIRGAS 2000 / UTM 23S - EPSG 31983
wkt_epsg31983 = """
Polygon (
    (
        192123.21329834562493488 8275487.08379946742206812,
        192325.92970819567563012 8275532.75235937628895044,
        192577.98180833790684119 8275500.78145875222980976,
        192656.32207071257289499 8275345.23357007838785648,
        192468.45366753044072539 8275140.64191859401762486,
        192260.18251823488390073 8275130.26117642782628536,
        192019.13764823798555881 8275288.66663362085819244,
        192123.21329834562493488 8275487.08379946742206812)
)
"""
sr_epsg31983 = osr.SpatialReference()
sr_epsg31983.ImportFromEPSG(31983)

imovel = ogr.CreateGeometryFromWkt(wkt_epsg31983, sr_epsg31983 ) # Geometry
###

# Mapa
layer_imovel = createLayerForLeaflet( imovel )
( layer_square, center ) = createBrasiliaSquareForLeaflet()

m = Map( center=center, zoom=10 )
m.add( layer_square )
m.add( layer_imovel )
m
```
